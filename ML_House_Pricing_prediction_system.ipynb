```
# stage 4 : Data transformation
# a. standerdization =[StanderedScaler] (x-mean)/ X_deviation --> [-1 to 1]
''' In statistics, standardization is the process of putting 
    different variables on the same scale. This process allows you to compare 
    scores between different types of variables. Typically, to standardize 
    variables, you calculate the mean and standard deviation for a variable. 
    Then, for each observed value of the variable, you subtract the mean 
    and divide by the standard deviation.'''

# calculate the mean of variable
X=df.price
Xmean=X.mean()
Xmean
# Calculate standard Deviation of variable
Xstd=X.std()
print('standered dev. of price is',Xstd)
# Standerdie the variable
scaledX= (X-Xmean)/Xstd
scaledX
```

```
scaledX
```

```
# b. Normalization =[minMaxScaler] (x-x_min)/(x_max-x_min) --> [0-1]
#      X = (X- Xmin)/(Xmax-Xmin)
 
# calculate the min value of x
Xmin=X.min()
Xmin

# calculate the max value of x
Xmax=X.max()
Xmax

# Feature SCALE THE PRICE VARIABLE
normalizedX = (X-Xmin)/(Xmax-Xmin)
normalizedX
```
